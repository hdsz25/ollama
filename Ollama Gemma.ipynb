```python
from langchain_openai import ChatOpenAI
from openai import AsyncOpenAI
import asyncio
import nest_asyncio
nest_asyncio.apply()

local_client= AsyncOpenAI(api_key="ollama",base_url='http://127.0.0.1:11434/v1')

query="请问你是谁？请详细介绍一下"
local_Async_settings={
    "model": "gemma:7b",
    # "model": "qwen:4b",
    # "model": "llama2:7b",
    "temperature": 0.7,
    "max_tokens": 1000,
    "top_p": 1,
    "frequency_penalty": 0,
    "presence_penalty": 0,
}
async def main(query):
    stream = await local_client.chat.completions.create(
        messages=[{"role": "user", "content": query}], stream=True, **local_Async_settings
    )

    async for part in stream:
        if token := part.choices[0].delta.content or "":
            print(token,end='')

```

```python
query="""Who is Grigori Perelman and why is he so important in mathematics?
(Answer with markdown sections, markdown with be the GitHub flavor.)"""
# query='你是谁？'

asyncio.run(main(query))
```
